PROCESO ETL

```python
# Importamos las librerias necesarias para realizar el proceso ETL
import pandas as pd
```

```python
# Importamos libreria para trabajar en Google Colaboratory
from google.colab import drive
drive.mount("/content/drive")
```

```python
# Cargamos los archivos con la información correspondiente a las plataformas de streaming
dfam=pd.read_csv("/content/drive/MyDrive/Proyecto_Lab1/amazon_prime_titles.csv")
dfdi=pd.read_csv("/content/drive/MyDrive/Proyecto_Lab1/disney_plus_titles.csv")
dfhu=pd.read_csv("/content/drive/MyDrive/Proyecto_Lab1/hulu_titles.csv")
dfne=pd.read_csv("/content/drive/MyDrive/Proyecto_Lab1/netflix_titles.csv")
```

Requerimiento 1: Generar campo id: Cada id se compondrá de la primera letra del nombre de la plataforma, seguido del show_id ya presente en los datasets (ejemplo para títulos de Amazon = as123)

```python
# Se realiza la codificación correspondiente
dfam.insert(loc=0, column="id", value="a" + dfam["show_id"])
dfdi.insert(loc=0, column="id", value="d" + dfdi["show_id"])
dfhu.insert(loc=0, column="id", value="h" + dfhu["show_id"])
dfne.insert(loc=0, column="id", value="n" + dfne["show_id"])
```

```python
# Se verifica la actualización de la información
dfne.head()
```

```python
# Con la actualización realizada se integran todos los archivos con la información de las plataformas
Total_df = pd.concat([dfam,dfdi,dfhu,dfne], axis=0)
Total_df.shape
```

Requerimiento 2: Los valores nulos del campo rating se deben reemplazar por el string “ G” (corresponde al rating de madurez: “general para todos los públicos”

```python
#Revisamos los valores nulos del campo rating
Total_df["rating"].isnull().sum()
```

```python
Total_df.info()
```

```python
#Realizamos el reemplazo correspondiente de los valores nulos por el string G
Total_df["rating"].fillna("G", inplace=True)
```

```python
# Se verifica la actualización de la información
Total_df.info()
```

Reqquerimiento 3: De haber fechas, debe tener el formato AAAA-mm-dd

```python
Total_df.head()
```

```python
# Damos el formato correspondiente al campo date_added
Total_df["date_added"] = pd.to_datetime(Total_df["date_added"])
```

```python
#Se verifica la actualización de la información
Total_df.head()
```

Requerimiento 4: Los campos de texto deben estar en minúsculas , sin excepciones

```python
Total_df.info()
```

```python
#Convertimos a minúscula los textos de los campos
Total_df["type"] = Total_df["type"].str.lower()
Total_df["title"] = Total_df["title"].str.lower()
Total_df["director"] = Total_df["director"].str.lower()
Total_df["cast"] = Total_df["cast"].str.lower()
Total_df["country"] = Total_df["country"].str.lower()
Total_df["rating"] = Total_df["rating"].str.lower()
Total_df["duration"] = Total_df["duration"].str.lower()
Total_df["listed_in"] =Total_df["listed_in"].str.lower()
Total_df["description"] = Total_df["description"].str.lower()
```

```python
Total_df.tail()
```

Requerimiento 5: El campo duracion debe convertirse en dos campos: duration_int y duration_type. El primero será un entero y el segundo una cadena indicando la unidad de medición de duración: min (minutos) o temporada (temporadas)

```python
# Se reliza la separación en 2 campos
splitcolumns = Total_df["duration"].str.split(" ", n=1, expand=True)
```

```python
# Se inserta una columna con la cantidad
Total_df["duration_int"] = splitcolumns[0]
```

```python
# Se inserta una columna con la unidad de medición
Total_df["duration_type"] = splitcolumns[1]
```

```python
# Se reemplaza "seasons" por "season" en la columna "duration_type"
Total_df["duration_type"] = Total_df["duration_type"].str.replace("seasons", "season")
```

```python
# Se cambia el formato "duration_int" a integer
Total_df["duration_int"] = Total_df["duration_int"].astype("Int64")
```

```python
# Se verifica la actualización de la información
Total_df.tail()
```

```python
# Se borran las columnas que ya no agreguen valor
Total_df= Total_df.drop(columns=['show_id','duration'])
Total_df.head()
```

```python
#Se revisa el archivo completo
Total_df.shape
```

```python
# Se realiza copia del archivo completo para su posterior utilización
Total_df.to_csv('/content/drive/MyDrive/Proyecto_Lab1/TotalStream.csv',index=False)
```
